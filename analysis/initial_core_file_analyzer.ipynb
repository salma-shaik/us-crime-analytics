### Separate rows with even 1 negative value to fix those records.

```python
initial_core_neg_nans = initial_core_df[initial_core_df < 0]

# drop records with nans in any non-numeric columns
initial_core_neg = initial_core_neg_nans.dropna(how='all', subset=list(initial_core_neg_nans)[8:])

initial_core_neg.to_csv('/Users/salma/Research/us-crime-analytics/data/analysis/initial_core_neg.csv',
                       index=False)
```

### Now get only the non-zero records i.e records with no negative value in any of the cells

```python
initial_core_pos = initial_core_df[initial_core_df > 0]
initial_core_pos.to_csv('/Users/salma/Research/us-crime-analytics/data/analysis/initial_core_pos.csv',
                       index=False)
```

### Another way to get only neg value records

```python
initial_core_neg = initial_core_df.merge(initial_core_pos, on=['ORI','YEAR'], 
                   how='left', indicator=True)

initial_core_neg_mgd = initial_core_neg[initial_core_neg['_merge'] == 'left_only']
initial_core_neg_mgd.to_csv('/Users/salma/Research/us-crime-analytics/data/analysis/initial_core_neg_test2.csv',
                                                                  index=False)
```
